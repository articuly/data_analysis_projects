```python
def get_group(x):
    if x in cluster1.index:
        return 'low_price'
    elif x in cluster2.index:
        return 'high_price_low_freq'
    else:
        return 'high_price_high_freq'

sf['group']=sf.neighborhood.apply(get_group)
```

## 特征工程

```python
sf.head()
```

```python
sf.drop(sf.columns[[0, 4, 6, 7, 8, 13]], axis=1, inplace=True)
sf.head()
```

```python
sf = sf[['bathrooms', 'bedrooms', 'finishedsqft', 'totalrooms', 'usecode', 'yearbuilt', 
         'zindexvalue', 'group', 'lastsoldprice']]
sf.head()
```

```python
#开始制作机器学习所需要的x和y 两个数据集
X=sf.iloc[:, :-1]
y=sf.iloc[:, -1:]
n=pd.get_dummies(sf['group'])
X=pd.concat([X, n], axis=1)
X.head()
```

```python
m=pd.get_dummies(sf['usecode'])
X=pd.concat([X, m], axis=1)
```

```python
drops=['group', 'usecode']
X.drop(drops, inplace=True, axis=1)
X.head()
```

```python
X.shape
```

```python
#标准化数据
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
X
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
print('linear regression R squared: {0:.4f}'.format(lr.score(X_test, y_test)))
```

在进行回归模型的评估是 R2是一个非常重要的指标 同时配合常见的MAE RMSE直接进行同步观察

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
lin_mse=mean_squared_error(y_pred, y_test)
lin_rmse=np.sqrt(lin_mse)
print('linear regression RMSE: {:.4f}'.format(lin_rmse))
```

```python
lin_mae=mean_absolute_error(y_pred, y_test)
print(f'linear regression MAE: {lin_mae:.4f}')
```

```python
# 使用随机森林来进行预测回归
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(random_state=42)
rfr.fit(X_train, y_train)
```

```python
print('random forest R squared: %.4f' %rfr.score(X_test, y_test))
```

```python
y_pred=rfr.predict(X_test)
rfr_mse=mean_squared_error(y_pred, y_test)
rfr_rmse=np.sqrt(rfr_mse)
print('random forest RMSE: %.4f' %rfr_rmse) 
```

```python
rfr_mae=mean_absolute_error(y_pred, y_test)
print(f'random forest MAE: {rfr_mae:.4f}')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor()
gbr.fit(X_train, y_train)
```

```python
print('gradient boosting R squared: %.4f' %gbr.score(X_test, y_test))
```

```python
y_pred=gbr.predict(X_test)
gbr_mse=mean_squared_error(y_pred, y_test)
gbr_rmse=np.sqrt(gbr_mse)
print('gradient boosting RMSE: %.4f' %gbr_rmse) 
```

```python
gbr_mae=mean_absolute_error(y_pred, y_test)
print(f'gradient boosting MAE: {gbr_mae:.4f}')
```
